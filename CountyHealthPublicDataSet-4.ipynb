### Creating a chart

Now that we have the numbers calculated, we want to show our findings in a chart.

Steps:

1.1 Begin by naming your subset something simple such as `NC_Data`. You will specify that the state you want included is North Carolina by renaming state as NC.

1.2 Next, you will combine the three subsets we created into one by including them all in one coding line.

1.3 Next, you will name the x-axis whatever you would like, in this case I named it metrics. Then I named those three metrics to what I would like them to be in the final chart. For the y-axis I will do the same and name it values and include the specific names of the data subsets we created previosuly.

1.4 You are now going to name the axies, add a title, add colors to the bars, and add in the figure size and limits.

- This should now produce a chart with all of the data collected, it will however be somewhat difficult to see the other two data bars as they are low in comparison to the median household income.

```python
NC_data = df[df["State"] == "NC"]
```

```python
median_income = NC_data["Median household income"].median()
poor_fair_health_percentage = (NC_data["Poor or fair health"].sum() / len(NC_data["Poor or fair health"])) * 100
uninsured_percentage = (NC_data["Uninsured adults"].sum() / len(NC_data["Uninsured adults"])) * 100
```

```python
metrics = ["Median Income", "Poor/Fair Health (%)", "Uninsured Adults (%)"]
values = [median_income, poor_fair_health_percentage, uninsured_percentage]
```

```python
plt.figure(figsize=(8, 6))
plt.bar(metrics, values, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title("Key Health and Economic Indicators for NC")
plt.ylabel("Value")
plt.ylim(0, max(values) * 1.1)
```

### Exporting our New Subsets

Once we've finished making our data subsets and creating more **useful** subsets for deeper analysis, we can export them as new .csv files, giving us the ability to reuse these datasets and further review them and have them available for others to view and analyze as well when posted on our public GitHub website.

#### Exporting to .csv file

To do so we can use the code `.to_csv()` - adding the filename and extension within the parentheses at the end.

For example, for our uninsured subset would put: `Uninsured_adults.to_csv("Uninsured_Adults.csv")` this will export a `.csv` file in our folder.

However, we only want to include the column names not the index numbers when exporting.

To eliminate the index number, we can add `index=false` to our line of code, which tells it not to bring in those index numbers.

`Uninsured_Adults.to_csv("Final.csv", index=False)`

We will repeat this for every data subset we would want to export and upload to GitHub.
